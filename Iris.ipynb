Here are examples of how to perform a t-test, ANOVA test, and chi-square test using the `scipy.stats` library and your Iris DataFrame.

**T-test**

A t-test is used to determine if there is a significant difference between the means of two groups. Let's perform an independent samples t-test to compare the sepal width of two species (e.g., Iris-setosa and Iris-versicolor).

```python
from scipy import stats

# Filter data for two species
setosa_sepal_width = df[df['species'] == 0]['sepal_width']
versicolor_sepal_width = df[df['species'] == 1]['sepal_width']

# Perform independent samples t-test
ttest_result = stats.ttest_ind(setosa_sepal_width, versicolor_sepal_width)

print(f"T-test statistic: {ttest_result.statistic}")
print(f"P-value: {ttest_result.pvalue}")

# Interpret the result
alpha = 0.05
if ttest_result.pvalue < alpha:
    print("Reject the null hypothesis: There is a significant difference in sepal width between the two species.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in sepal width between the two species.")
```

**ANOVA Test**

ANOVA (Analysis of Variance) is used to compare the means of three or more groups. Let's perform a one-way ANOVA to compare the sepal length across all three species.

```python
# Separate sepal length data for each species
setosa_sepal_length = df[df['species'] == 0]['sepal_length']
versicolor_sepal_length = df[df['species'] == 1]['sepal_length']
virginica_sepal_length = df[df['species'] == 2]['sepal_length']

# Perform one-way ANOVA test
anova_result = stats.f_oneway(setosa_sepal_length, versicolor_sepal_length, virginica_sepal_length)

print(f"ANOVA F-statistic: {anova_result.statistic}")
print(f"P-value: {anova_result.pvalue}")

# Interpret the result
alpha = 0.05
if anova_result.pvalue < alpha:
    print("Reject the null hypothesis: There is a significant difference in sepal length across the three species.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in sepal length across the three species.")
```

**Chi-Square Test**

A chi-square test is used to examine the relationship between two categorical variables. While all variables in your current dataframe are numerical after encoding, we can demonstrate a chi-square test by creating a hypothetical categorical variable based on one of the continuous variables, or by using the original 'species' column and a hypothetical categorical variable.

Let's create a categorical variable for 'sepal_width' by binning it and then perform a chi-square test between this new variable and the 'species' variable (which is treated as categorical after encoding).

```python
# Create a categorical variable by binning sepal_width
df['sepal_width_category'] = pd.cut(df['sepal_width'], bins=3, labels=['narrow', 'medium', 'wide'])

# Create a contingency table of the two categorical variables
contingency_table = pd.crosstab(df['species'], df['sepal_width_category'])
print("Contingency Table:")
display(contingency_table)

# Perform the chi-square test
chi2_result = stats.chi2_contingency(contingency_table)

print(f"\nChi-square statistic: {chi2_result.statistic}")
print(f"P-value: {chi2_result.pvalue}")
print(f"Degrees of freedom: {chi2_result.dof}")
print(f"Expected frequencies:\n{chi2_result.expected_freq}")

# Interpret the result
alpha = 0.05
if chi2_result.pvalue < alpha:
    print("Reject the null hypothesis: There is a significant association between species and sepal width category.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between species and sepal width category.")
```
